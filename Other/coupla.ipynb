```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sigma=np.array([[1,-.8],[-.8,1]])
alphas=np.array([3,2])
betas=np.array([5,5])
N=1000
```

```python
u=np.random.multivariate_normal((0,0),sigma,size=N)
c=sp.stats.norm.cdf(u)
generated=sp.stats.gamma.ppf(c,alphas,betas)
df = pd.DataFrame(list(zip(generated[:, 0], generated[:, 1])),
               columns =['x1', 'x2'])
```

```python
sns.jointplot(df, x="x1", y="x2", kind='kde')
plt.show()
```
